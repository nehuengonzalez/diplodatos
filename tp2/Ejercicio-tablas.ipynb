```python
filtrado= tabla_1[tabla_1.Sex==1]
filtrado.plot(x='Agemos', y='P3');
```

```python
for y in ['P3', 'P5', 'P10', 'P25', 'P50', 'P75','P90', 'P95', 'P97']:
    seaborn.lineplot(data=filtrado, x='Agemos', y=y)
    
seaborn.despine()
plt.title('Pesos Masculino')
plt.legend(labels = ['P3', 'P5', 'P10', 'P25', 'P50', 'P75','P90', 'P95', 'P97'], ncol=2, loc = 4)
```

```python
for y in ['P3', 'P5', 'P10', 'P25', 'P50', 'P75','P90', 'P95', 'P97']:
    seaborn.lineplot(data=tabla_1[tabla_1.Sex == 2], x='Agemos', y=y)
    
seaborn.despine()
plt.title('Pesos Femenino')
plt.legend(labels = ['P3', 'P5', 'P10', 'P25', 'P50', 'P75','P90', 'P95', 'P97'], ncol=2, loc=4)
```

```python
def convert(x):
    try :
        return float(x)
    except:
        return np.nan
    
tabla_2 = pd.read_csv("./input/lenageinf.csv", converters={ i:convert for i in range(100)})
tabla_2[30:50]
```

```python
for y in ['P3', 'P5', 'P10', 'P25', 'P50', 'P75','P90', 'P95', 'P97']:
    seaborn.lineplot(data=tabla_2[tabla_2.Sex == 1], x='Agemos', y=y)
    
seaborn.despine()
plt.title('Longitud Masculino')
plt.legend(labels = ['P3', 'P5', 'P10', 'P25', 'P50', 'P75','P90', 'P95', 'P97'], ncol=2, loc=4)
```

```python
for y in ['P3', 'P5', 'P10', 'P25', 'P50', 'P75','P90', 'P95', 'P97']:
    seaborn.lineplot(data=tabla_2[tabla_2.Sex == 2], x='Agemos', y=y)
    
seaborn.despine()
plt.title('Longitud Femenino')
plt.legend(labels = ['P3', 'P5', 'P10', 'P25', 'P50', 'P75','P90', 'P95', 'P97'], ncol=2, loc=4)
```
